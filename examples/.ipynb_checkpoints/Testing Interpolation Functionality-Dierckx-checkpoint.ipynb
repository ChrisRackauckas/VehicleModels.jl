# Using Dierckx  - -- > creating a spline

```julia
using Dierckx
using Plots
```

```julia
x1 = Float64
y1 = Float64
x = [0., 1., 2., 3., 4., 5.3]
y = [-1., 0., 7., 26., 63., 20]  # x.^3 - 1.
spl_1D = Spline1D(x,y,k=1)    # linear spline
x1 = 4.2
y1 = spl_1D(x1)  # careful with the (x,y) order!
plot(x,y,label="data")
scatter!((x1, y1), markershape = :hexagon, markersize = 12, label="1D point",leg=true)
```
