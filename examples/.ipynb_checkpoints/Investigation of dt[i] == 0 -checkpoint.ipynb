# So what should be done?
The states are changing when no time elapses, but since there is not limit placed on them they can do that. Do we:
    1) ignore p2 and jump straight to p3
    2) change the formulation and add limits on the rates of ALL of the states
    3) change the formulation to make sure that these dt[i] == 0 solutions are all at the end of the simulation
        * How would we accomplish this?
The integrals of the states that are changing with no time elapsed are not able to change during this time interval. For instance; 
    0 == sa[i+1]  - sa[i]     - sr[i]*dt[i]
If dt[i] == 0, then sa[i+1] == sa[i]
